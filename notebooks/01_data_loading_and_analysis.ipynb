# Part 1: Data Loading and Analysis

## 1. Data Loading

### 1.1 Load Data from ArcGIS

```python
import requests

url = "https://maps.ottawa.ca/arcgis/rest/services/WaterandWastewaterInfrastructure/MapServer/8/query"

pipes = []

offset = 0
batch_size = 1000

while True:
    params = {
        "where": "1=1",
        "outFields": "*",
        "returnGeometry": "true",
        "resultOffset": offset,
        "resultRecordCount": batch_size,
        "f": "json"
    }

    data = requests.get(url, params=params).json()

    features = data.get("features", [])

    if len(features) == 0:
        break
    pipes.extend(features)


    offset += batch_size
pipes
```

### 1.2 Export Raw Data

```python
import json
with open("data/pipes_raw.json", "w") as file:
    json.dump(pipes, file, indent=4)
```

### 1.3 Make Data into a Table

```python
import pandas as pd
pipes_a = pd.DataFrame([object["attributes"] for object in pipes])
pipes_gx1 = pd.DataFrame([object["geometry"]["paths"][0][0][0] for object in pipes], columns=["x1"])
pipes_gy1 = pd.DataFrame([object["geometry"]["paths"][0][0][1] for object in pipes], columns=["y1"])
pipes_gx2 = pd.DataFrame([object["geometry"]["paths"][0][1][0] for object in pipes], columns=["x2"])
pipes_gy2 = pd.DataFrame([object["geometry"]["paths"][0][1][1] for object in pipes], columns=["y2"])
```

```python
pipes = pd.concat([pipes_a, pipes_gx1, pipes_gy1, pipes_gx2, pipes_gy2], axis=1)
pipes
```

### 1.4 Export Tabular Data

```python
pipes.to_csv("data/pipes_tabular.csv", index=False)
```

## 2. Basic Data Cleaning

```python
pipes.info()
```

The following columns will be removed
- OBJECTID, SUBTYPE, STRUCT_ID, GLOBAL_ID, SHAPE_Length are unhelpful features
- STRUCT_TYPE and LIFECYCLE_STATUS all consist of only one unique value
- ENABLED is all 1 except 1 sample
- 42% of FUNCTION are NaN

```python
pipes_cleaned = pipes.drop(["OBJECTID", "SUBTYPE", "STRUCT_ID", "GLOBALID", "STRUCT_TYPE", "LIFE_CYCLE_STATUS", "ENABLED", "FUNCTION", "SHAPE_Length"], axis=1)
pipes_cleaned
```

```python
pipes_cleaned = pipes_cleaned.dropna()
pipes_cleaned
```

```python
import numpy as np

cols_to_check = ["INSTALL_YEAR", "WIDTH", "LENGTHASBUILT", "x1", "y1", "x2", "y2"]
pipes_cleaned[cols_to_check] = pipes_cleaned[cols_to_check].replace(0.0, np.nan)
```

```python
pipes_cleaned = pipes_cleaned.dropna()
```
